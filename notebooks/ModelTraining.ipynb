```python
from ultralytics import YOLO
```

```python
import torch
print(torch.cuda.is_available())
```

```python
model1 = YOLO("yolo26n.pt")

results1 = model1.train(
    data="/content/drive/MyDrive/GRP19_SA1/Annotated Datasets/SA01_Annotations.v2-final.yolov8/data.yaml",
    epochs=25,
    imgsz=640,
    batch=4,
    optimizer="AdamW",
    lr0=0.01
)
```

```python
model2 = YOLO("yolo26n.pt")

results2 = model2.train(
    data="/content/drive/MyDrive/GRP19_SA1/Annotated Datasets/SA01_Annotations.v2-final.yolov8/data.yaml",
    epochs=30,
    imgsz=640,
    batch=20,
    optimizer="SGD",
    lr0=0.001
)
```

```python
model3 = YOLO("yolo26n.pt")

results3 = model3.train(
    data="/content/drive/MyDrive/GRP19_SA1/Annotated Datasets/SA01_Annotations.v2-final.yolov8/data.yaml",
    epochs=40,
    imgsz=640,
    batch=-1,
    optimizer="auto",
    lr0=0.0001
)
```

```python
metrics1 = model1.val()
metrics2 = model2.val()
metrics3 = model3.val()
```

```python
import pandas as pd

hyper_table = pd.DataFrame({
    "Model": ["Model 1", "Model 2", "Model 3"],
    "Epochs": [25, 30, 40],
    "Optimizer": ["AdamW", "SGD", "auto"],
    "Batch Size": [4, 20, -1],
    "Learning Rate": [0.01, 0.001, 0.0001],
    "Image Size": [640, 640, 640]
})

hyper_table
```

```python
metrics_table = pd.DataFrame({
    "Model": ["Model 1", "Model 2", "Model 3"],
    "mAP50": [
        metrics1.box.map50,
        metrics2.box.map50,
        metrics3.box.map50
    ],
    "Precision": [
        metrics1.box.mp,
        metrics2.box.mp,
        metrics3.box.mp
    ],
    "Recall": [
        metrics1.box.mr,
        metrics2.box.mr,
        metrics3.box.mr
    ],
    "F1 Score": [
        2*(metrics1.box.mp*metrics1.box.mr)/(metrics1.box.mp+metrics1.box.mr),
        2*(metrics2.box.mp*metrics2.box.mr)/(metrics2.box.mp+metrics2.box.mr),
        2*(metrics3.box.mp*metrics3.box.mr)/(metrics3.box.mp+metrics3.box.mr)
    ]
})

metrics_table
```
