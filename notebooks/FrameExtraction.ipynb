```python
#TSet up where system will read and save files
import os

Base_PATH = "/content/drive/MyDrive/GRP19_SA01"
RawVideo_PATH = os.path.join(Base_PATH, "Raw Videos")
FrameOutput_PATH = os.path.join(Base_PATH, "Extracted Frames-1FPS")

os.makedirs(FrameOutput_PATH, exist_ok=True)

print("Base Path:", Base_PATH)
print("Raw Videos Path:", RawVideo_PATH)
print("Output Folder Created:", FrameOutput_PATH)
```

```python
import cv2
import os

def extract_frames(video_path, output_folder, fps_rate=1):
    VideoName = os.path.basename(video_path).split('.')[0]
    VideoOutput = os.path.join(output_folder, VideoName)
    os.makedirs(VideoOutput, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    frameIntervals = int(fps / fps_rate)

    frameCount = 0
    savedCount = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if frameCount % frameIntervals == 0:
            FrameFile = os.path.join(
                VideoOutput,
                f"{VideoName}_frame_{savedCount:05d}.jpg"
            )
            cv2.imwrite(FrameFile, frame)
            savedCount += 1

        frameCount += 1

    cap.release()
    print(f"Finished {VideoName} → Saved {savedCount} frames")
```

```python
VideoFiles = [f for f in os.listdir(RawVideo_PATH)
               if f.lower().endswith((".mp4", ".avi"))]
print("Videos detected:", VideoFiles)

for video in VideoFiles:
    Video_PATH = os.path.join(RawVideo_PATH, video)
    extract_frames(Video_PATH, FrameOutput_PATH, fps_rate=1)
```
